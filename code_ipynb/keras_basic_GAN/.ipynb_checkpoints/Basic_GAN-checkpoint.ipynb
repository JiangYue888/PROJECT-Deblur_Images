## Testing model

```python
import numpy as np
from PIL import Image
from model import generator_model
from utils import load_image, load_images,deprocess_image, preprocess_image
import math
```

```python
def PSNR(img1, img2):
    ##Peak signal to noise ratio
    ##Evaluating sharp and deblurred images
    mse = np.mean( (img1/255. - img2/255.) ** 2 )
    if mse == 0:
        return 100
    PIXEL_MAX = 1
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
def test_visual(batch_size):
    for i in range(generated_images.shape[0]):
        y = y_test[i, :, :, :]
        x = x_test[i, :, :, :]
        img = generated[i, :, :, :]
        output = np.concatenate((y, x, img), axis=1)
        im = Image.fromarray(output.astype(np.uint8))
        im.save('visual_evaluation/results{}.png'.format(i))
def test_show(ind):
    y = y_test[ind, :, :, :]
    x = x_test[ind, :, :, :]
    img = generated[ind, :, :, :]
    output = np.concatenate((y, x, img), axis=1)
    im = Image.fromarray(output.astype(np.uint8))
    return(im)
```

Evaluation Metrics: **PSNR** (Peak Signal to Noise Ratio)

Greater value means better deblur result


```python
##set size of testing data
batch_size=100
##load data (images)
data = load_images('./images/train',-1)
```

```python
##set size of testing data
batch_size=100
##load data (images)
data = load_images('./images/test', batch_size)
y_test, x_test = data['B'], data['A']
##generate images
g = generator_model()
g.load_weights('generator.h5')##Using original model 
generated_images = g.predict(x=x_test, batch_size=batch_size)
generated = np.array([deprocess_image(img) for img in generated_images])
x_test = deprocess_image(x_test)
y_test = deprocess_image(y_test)
```

```python
##Mean PSNR of original model
psnr=0
for i in range(generated_images.shape[0]):
        y = y_test[i, :, :, :]
        img = generated[i, :, :, :]
        psnr=psnr+PSNR(y,img)
        psnr_mean=psnr/batch_size
psnr_mean
```
